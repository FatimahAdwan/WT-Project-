```python
# set ups
import pandas as pd
pd.plotting.deregister_matplotlib_converters
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

```

in this project, we will use a dataset from kaggle, the dataset is on PC (personal computer), it is a game played on PC rather than video game console or arcade machine and how it trends over a period of time to present date. 

```python
# importing and reading the data
computer_games = pd.read_csv("../input/computer-games-dataset/computer_games.csv")
computer_games.head()
```

```python
# checking to see for the types data in here colloctively
computer_games.info()
```

```python
# also checking for null values 
computer_games.isna().value_counts()
```

```python
# statistical summary of the data set
computer_games.describe()
```

```python
# we would wanna check for duplicate just incase there is any and remove them
computer_games.drop_duplicates().describe()
```

```python
# we start by checking for the developers of the game
top_50_developers = computer_games["Developer"].value_counts().head(50)
top_50_developers
```

```python
# plt.figure(figsize = (16, 10))
top_50_developers.plot(kind = "barh", figsize = (10, 10),color = "green", title = "Top computer game developers" )
```

```python
# nxet we check for the top producers
top_50_producers = computer_games["Producer"].value_counts().head(50)
top_50_producers
```

```python
top_50_producers.plot(kind = "barh", figsize = (10,10), color ="green",title = "Top computer game producing companies" )
```

```python
top_50_genres = computer_games["Genre"].value_counts().head(50)
top_50_genres
```

```python
# make a plot
top_50_genres.plot(kind = "barh", figsize = (10,10), color = "green", title = "Top computer game genres")
```

insight that can be derived from here is that first person shooting games are mostly produced and played games, like Call of Duty, Battlefield, and Counter-Strike are all First person shooting games
